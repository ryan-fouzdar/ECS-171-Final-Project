```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv("Diabetes.csv")

# Additional distributions and box plots for Blood Pressure, Insulin, Age, and DiabetesPedigreeFunction

# Plot Blood Pressure distribution & Blood Pressure vs Diabetes outcome
sns.histplot(df['BloodPressure'], kde=True)
plt.title('Blood Pressure Distribution')
plt.show()
sns.boxplot(x='Outcome', y='BloodPressure', data=df)
plt.title('Blood Pressure vs Diabetes Outcome')
plt.show()

# Plot Insulin distribution & Insulin vs Diabetes outcome
sns.histplot(df['Insulin'], kde=True)
plt.title('Insulin Distribution')
plt.show()
sns.boxplot(x='Outcome', y='Insulin', data=df)
plt.title('Insulin vs Diabetes Outcome')
plt.show()

# Plot Age distribution & Age vs Diabetes outcome
sns.histplot(df['Age'], kde=True)
plt.title('Age Distribution')
plt.show()
sns.boxplot(x='Outcome', y='Age', data=df)
plt.title('Age vs Diabetes Outcome')
plt.show()

# Plot Diabetes Pedigree distribution & Diabetes Pedigree vs Diabetes outcome
sns.histplot(df['DiabetesPedigreeFunction'], kde=True)
plt.title('Diabetes Pedigree Function Distribution')
plt.show()
sns.boxplot(x='Outcome', y='DiabetesPedigreeFunction', data=df)
plt.title('Diabetes Pedigree Function vs Diabetes Outcome')
plt.show()
```
