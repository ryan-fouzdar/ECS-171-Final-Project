Pearson Correlation and Feature Heat Map:


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = pd.read_csv('/content/drive/MyDrive/Diabetes.csv') # replace file path on your machine

f1 = 'Glucose' # Modify features as needed, make sure to reflect changes in X assignment and "sns.heatmap" line
f2 = 'BloodPressure'
f3 = 'SkinThickness'
f4 = 'Insulin'
f5 = 'BMI'
f6 = 'Age'

X = data[[f1,f2,f3,f4,f5,f6]] # features being compared modify as needed
y = data['Outcome'] # can be modified to compare between model features and outcome

correlation_matrix = X.corr() # function to calculate Pearson Coefficient

plt.figure(figsize=(11, 9))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', xticklabels=[f1, f2,f3, f4, f5, f6], yticklabels=[f1,f2,f3,f4,f5,f6])
plt.title('Correlation Heatmap between Focus Features')
plt.show()
```
